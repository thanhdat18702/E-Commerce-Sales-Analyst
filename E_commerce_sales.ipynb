```python
import pandas as pd

df = pd.read_csv('/content/Global_Superstore2.csv', encoding='latin-1')
```

```python
df=df.dropna(how='all').reset_index(drop=True)
```

```python
# Xóa nhiều cột từ DataFrame df
df = df.drop(['Order ID', 'Customer ID', 'Customer Name', 'Postal Code', 'Product ID', 'Product Name'], axis=1)
```

```python
df.describe().T
```

```python
df['Order Date']=df['Order Date'].replace('-','/')
df['Ship Date']=df['Ship Date'].replace('-','/')
```

```python
df.to_csv('output_final.csv', encoding = 'utf-8-sig')
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pylab as mp
import pandas as pd # make sure pandas is imported

# Convert 'Ship Mode' to category and then to numeric codes
df['Ship Mode'] = df['Ship Mode'].astype('category').cat.codes

# Convert all columns to numeric type except for 'Order Date' and 'Ship Date'
for col in df.columns:
    if col not in ['Order Date', 'Ship Date']:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Exclude 'Order Date' and 'Ship Date' from the correlation matrix calculation
dataplot = sns.heatmap(df.drop(['Order Date', 'Ship Date'], axis=1).corr().round(1), cmap="YlGnBu", annot=True)
mp.show()
```
